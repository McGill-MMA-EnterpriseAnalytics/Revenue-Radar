```python
#Import dataset(s)
train_revenue = pd.read_csv("train.csv")
test_revenue = pd.read_csv("test.csv")
```

```python
dict_columns = ['geoNetwork', 'device', 'totals', 'trafficSource']

# For each such column, convert the string representation of dictionary to actual dictionary
# then normalize and concatenate with the original DataFrame
for column in dict_columns:
    # Convert the stringified dictionary to actual dictionary
    train_revenue[column] = train_revenue[column].apply(json.loads)
    
    # Flatten the dictionary and convert to a DataFrame
    column_df = pd.json_normalize(train_revenue[column])
    
    # Concatenate the new DataFrame to the original DataFrame
    train_revenue = pd.concat([train_revenue, column_df], axis=1)

    # Drop the original column that contained the stringified dictionary
    train_revenue.drop(column, axis=1, inplace=True)

for column in dict_columns:
    # Convert the stringified dictionary to actual dictionary
    test_revenue[column] = test_revenue[column].apply(json.loads)
    
    # Flatten the dictionary and convert to a DataFrame
    column_df = pd.json_normalize(test_revenue[column])
    
    # Concatenate the new DataFrame to the original DataFrame
    test_revenue = pd.concat([test_revenue, column_df], axis=1)

    # Drop the original column that contained the stringified dictionary
    test_revenue.drop(column, axis=1, inplace=True)
```

### Handling the Date Column
- Split the date column into year, month, and day.
- Dropped the year as it is not useful for predictive modeling, since past years won't resurface unless we are conducting some temporal analysis.
- Bucketed the months into first, second, and third quarters.
- Bucketed the days into beginning, middle, and end of the month.
- Created a new feature to capture whther a visit was on a weekend or not

```python
# splitting the date column into year, month and day
df['month']=df['date'].dt.month
df['day']=df['date'].dt.day
df['year']=df['date'].dt.year
```

```python
df.head()
```

```python
# bucketing month into quarters
def quarter(x):
    if x in [1,2,3]:
        return '1st_quarter'
    elif x in [4,5,6]:
        return '2nd_quarter'
    elif x in [7,8,9]:
        return '3rd_quarter'
    else:
        return '4th_quarter'
    
df['quarter']=df['month'].apply(quarter)
```

```python
# bucking the day column into beginning, middle and end of the month
def day(x):
    if x in range(1,11):
        return 'beginning'
    elif x in range(11,21):
        return 'middle'
    else:
        return 'end'

df['day_of_month']=df['day'].apply(day)
```

```python
# Creating a new column 'is_weekend' which will indicate whether the day is a weekday or a weekend
df['is_weekend'] = df['date'].dt.dayofweek.apply(lambda x: 1 if x >= 5 else 0)
df['is_weekend'].value_counts()
```
